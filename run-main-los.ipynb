```python

import pandas as pd
import matplotlib.pyplot as plt
##compare all test result 
dataset = 'los'
path = f"out/tgcn/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre1_epoch150_adj-dist/test_result.xlsx"
data = pd.read_excel(path, sheet_name='pred', header=None, engine='openpyxl')
test_result = data.values
data = pd.read_excel(path, sheet_name='true', header=None, engine='openpyxl')
true_labels = data.values

path = f'out/tgcn/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre1_epoch150_adj-gsl/test_result.xlsx'
data = pd.read_excel(path, sheet_name='pred', header=None, engine='openpyxl')
test_result_gsl = data.values
# data = pd.read_excel(path, sheet_name='true', header=None)
# true_labels = data.values
path = f'out/tgcn/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre1_epoch150_adj-gsldist/test_result.xlsx'
data = pd.read_excel(path, sheet_name='pred', header=None, engine='openpyxl')
test_result_gsldist = data.values

path = 'out/tgcn'
fig1 = plt.figure(figsize=(15,5))
#    ax1 = fig1.add_subplot(1,1,1)
a_pred = test_result[:,0]
a_pred_gsl = test_result_gsl[:,0]
a_pred_gsldist = test_result_gsldist[:,0]
a_true = true_labels[:,0]
plt.plot(a_true,'b-',label='Ground Truth', linewidth=1.0)
plt.plot(a_pred,'r--',label='T-GCN', linewidth=2.0)
plt.plot(a_pred_gsl,'g-',label='T-GCN-GSL', linewidth=3.0)
plt.plot(a_pred_gsldist,'k:',label='T-GCN-GSL-DIST', linewidth=2.0)
plt.legend(loc='best',fontsize=15)
file_name = f'/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre1_epoch150_adj_compare_all.jpg'
plt.savefig(path+file_name)
plt.show()
## oneday test result visualization
fig1 = plt.figure(figsize=(17,5))
#    ax1 = fig1.add_subplot(1,1,1)
a_pred = test_result[:96,0]
a_pred_gsl = test_result_gsl[:96,0]
a_pred_gsldist = test_result_gsldist[:96,0]
a_true = true_labels[0:96,0]
plt.plot(a_true,'b-',label='Ground Truth', linewidth=1.0)
plt.plot(a_pred,'r--',label='T-GCN', linewidth=2.0)
plt.plot(a_pred_gsl,'g-',label='T-GCN-GSL', linewidth=3.0)
plt.plot(a_pred_gsldist,'k:',label='T-GCN-GSL-DIST', linewidth=2.0)
plt.legend(loc='best',fontsize=15)
file_name = f'/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre1_epoch150_adj_test_oneday.jpg'
plt.savefig(path+file_name)
plt.show()
```
